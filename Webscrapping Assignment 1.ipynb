```python
########Ans 1##########
from bs4 import BeautifulSoup
import requests
def Header(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html5lib')
    heading_tags = ["h1", "h2", "h3"]
    table=soup.find_all(heading_tags)
    for i in table:
        print(i)
Header("https://en.wikipedia.org/wiki/Main_Page")
```

```python
#############Ans 2################
import re
import pandas as pd
def ImdbTop100(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html5lib')
    movies = soup.select('td.titleColumn')
    crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
    ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
    li = []

    for index in range(0, len(movies)):


        movie_string = movies[index].get_text()
        movie = (' '.join(movie_string.split()).replace('.', ''))
        movie_title = movie[len(str(index))+1:-7]
        year = re.search('\((.*?)\)', movie_string).group(1)
        place = movie[:len(str(index))-(len(movie))]
        data = {"place": place,
                "movie_title": movie_title,
                "rating": ratings[index],
                "year": year,
                "star_cast": crew[index],
                }
        li.append(data)
    for movie in li[:100]:
        print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
            ') -', 'Starring:', movie['star_cast'], movie['rating'])
    df = pd.DataFrame(li[:100])
ImdbTop100('http://www.imdb.com/chart/top')
```

```python
############Ans 3######################
def ImdbIndianTop100(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html5lib')
    results=soup.find_all("td")
    l=[]
    for i in range(1,len(results),5):
        if i==len(results)-1:
            break
        else:
            name=results[i].text.strip().split(".")[-1].strip().split("\n")[0]
            year=results[i].text.strip().split(".")[-1].strip().split("\n")[-1][9:13]
            rating=results[i+1].text.strip()
        l.append([name,year,rating])
        if len(l)==100:
            break
    for i in range(len(l)):
        print(str(i+1)+" "+l[i][0]+" "+l[i][1]+" "+l[i][2])
    df = pd.DataFrame(l)
    df.columns=["Movie Name","Year","IMDB Rating"]
    print(df)
            
ImdbIndianTop100("https://www.imdb.com/india/top-rated-indian-movies/?sort=ir,desc&mode=simple&page=1")   
```

```python
###########Ans 4####################
def President_of_India(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html5lib')
    temp=soup.find("ul",{'class':"listing cf"})
    results=temp.findAll("li")
    names=[]
    term=[]
    for i in range(len(results)):
        names.append(results[i].find("h3").text.split("(")[0])
        term.append(results[i].find("p").text.split("</span>")[0])
    for i in range(len(names)):
        print("Name: "+" "+names[i]+" "+term[i])
    
            
President_of_India("https://presidentofindia.nic.in/former-presidents.htm")   
```

```python
###########Ans 5####################
def ODIMenTeams(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html5lib')
    team=[]
    match=[]
    points=[]
    ratings=[]
    temp=soup.find("tr",{'class':"rankings-block__banner"})
    team1=temp.find("td",{'class':"rankings-block__banner--team-name"})
    team.append(team1.find("span",{'class':'u-hide-phablet'}).text)
    match.append(temp.find("td",{'rankings-block__banner--matches'}).text)
    points.append(temp.find("td",{'class':'rankings-block__banner--points'}).text)
    ratings.append(temp.find("td",{'class':'rankings-block__banner--rating u-text-right'}).text.strip())
    temp=soup.findAll("tr",{"class":"table-body"})
    for i in range(9):
        team1=temp[i].find("td",{'class':"table-body__cell rankings-table__team"})
        team.append(team1.find("span",{'class':'u-hide-phablet'}).text)
        match.append(temp[i].findAll("td",{'table-body__cell u-center-text'})[0].text)
        points.append(temp[i].findAll("td",{'class':'table-body__cell u-center-text'})[1].text)
        ratings.append(temp[i].find("td",{'class':'table-body__cell u-text-right rating'}).text.strip())
    for i in range(len(team)):
        print(team[i]+" "+match[i]+" "+points[i]+" "+ratings[i])

ODIMenTeams("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")  
```

```python
def Headline(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html5lib")
    temp=soup.find("ul",{"class":"LatestNews-list"})
    results=temp.findAll("li","LatestNews-item")
    headline=[]
    time=[]
    links=[]
    for i in range(len(results)):
        headline.append(results[i].find("a",{"class":"LatestNews-headline"}).text.strip())
        time.append(results[i].find("time",{"class":"LatestNews-timestamp"}).text.strip())
        l=soup.find("a",{"class":"LatestNews-headline"})
        links.append(l.get("href"))
    for i in range(len(headline)):
        print(headline[i]+"    "+time[i]+"    "+links[i])
    
Headline("https://www.cnbc.com/world/?region=world%20:")
```

```python
############Ans 9#############
def dineIn(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html5lib")
    result=soup.find("div",{"class":"restnt-card-wrap-new"})
    temp=result.findAll("div",{"class":"restnt-card restaurant"})
    rest=[]
    cuisine=[]
    c=[]
    loc=[]
    l=""
    ratings=[]
    imgurl=[]
    for i in range(len(temp)):
        rest.append(temp[i].find("div",{"class":"restnt-info cursor"}).text)
        cus=temp[i].find("span",{"class":"double-line-ellipsis"})
        for j in cus.findAll("a"):
            c.append(j.text)
        cuisine.append(c)
        c=[]
        lo=temp[i].find("div",{"class":"restnt-loc ellipsis"})
        for j in lo.findAll("a"):
            l+=j.text+", "
        loc.append(l)
        ratings.append(temp[i].find("div",{"class":"restnt-rating rating-4"}).text)
        im=temp[i].find('img',{"class":"no-img"})
        imgurl.append(im.get('data-src'))
        #l=soup.find("a","sc-5smygv-0 nrDZj")
        #links.append(l.get("href"))
    for i in range(len(rest)):
        print(rest[i])
        print(cuisine[i])
        print(loc[i])
        print(ratings[i])
        print(imgurl[i])
        print("\n")
    
dineIn("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
###########Ans 10############
def scholar(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html5lib")
    rank=list(soup.findAll("td",{"class":"gsc_mvt_p"}))
    publication=list(soup.findAll("td",{"class":"gsc_mvt_t"}))
    h5_index=list(soup.findAll("a",{"class":"gs_ibl gsc_mp_anchor"}))
    h5_median=list(soup.findAll("span",{"class":"gs_ibl gsc_mp_anchor"}))
    r=[]
    p=[]
    h5i=[]
    h5m=[]
    for i in range(len(rank)):
        r.append(rank[i].text)
        p.append(publication[i].text)
        h5i.append(h5_index[i].text)
        h5m.append(h5_median[i].text)
    for i in range(len(r)):
        print(str(r[i])+" "+str(p[i])+" "+str(h5i[i])+" "+str(h5m[i]))
    
scholar("https://scholar.google.com/citations?view_op=top_venues&hl=en")
```
